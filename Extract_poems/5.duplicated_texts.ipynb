# Elimiminate duplicated poems

Cosine distance revealed that some of the poems are duplicated. This notebook will identify and eliminate them.

## Result
It looked like duplicates but it turned out it was all poems shorter than a page that shared a page with another poem in the corpus. Since this version of the text does not have not been cleaned that means that the texts are duplicated, but thats not a mistake at this point.


```python
from pymongo import MongoClient
import json
import pandas as pd
from norn.poems import Poem

db = MongoClient().norn
collection = db['poems']
```

```python
# Find the duplicated texts

documents = collection.find() # Get all documents
df = pd.DataFrame(list(documents)) # Cast to df
duplicated = df.loc[df.duplicated(subset="text", keep=False)].sort_values(by="text")
```
