```python
import pymongo
from pathlib import Path
from dataclasses import dataclass
import os
from bson.objectid import ObjectId
from norn.poems import Poem
import pandas as pd

poems = Path('/mnt/md1/new_2023/norn/reocr_poems2/')

client = pymongo.MongoClient()

collection = client['norn']['poems']
```

```python
# Create a dataframe with the poems

@dataclass
class PoemText:
    urn: str
    # title: str
    dhlabid: str
    text: str   
```

```python
# write poems to mongodb
for path, _, file in os.walk(poems):
    if file:
        for f in file:
            with open(path + '/' + f, 'r') as poem:
                text = poem.read()
                urn = path.split('/')[-1]
                dhlabid = f.split('.')[0]
                poem_text = PoemText(urn, dhlabid, text)
                collection.insert_one(poem_text.__dict__)
    
```

## Add metadata from sheet

```python
docs = collection.find()
```

```python
df = pd.DataFrame(docs)

df
```

```python
metadata = pd.read_csv(Path('..') / "Extract_poems" / "poems.csv")

metadata
```

```python
merged = pd.merge(df, metadata, on=['dhlabid', 'urn'])

merged.head(5)
```

```python
# Update database with metadata
# Title, page_start, page_end, overlapp, digital_visning, comment

for index, row in merged.iterrows():
    query = {'_id': ObjectId(row['_id'])}
    update = {'$set': {'title': row['title'], 'page_start': row['page_start'], 'page_end': row['page_end'], 'overlapp': row['overlapp'], 'digital_visning': row['digital_visning'], 'comment': row['comment']}}
    
    collection.update_one(query, update)
```

```python
outer = merged.loc[:, ["_id", "urn"]].merge(df, on="urn", how="left").drop_duplicates()
```

```python
outer.drop_duplicates(subset=['_id'], inplace=True)
```

```python
outer.drop(columns=['publisher', 'publisher_place'], inplace=True)
```

```python
# Update database with metadata
# author, title, year

for index, row in outer.iterrows():
    query = {'_id': ObjectId(row['_id'])}
    update = {'$set': {'author': row['author'],  'book_title': row['title'], 'year': row['year']}}
    
    collection.update_one(query, update)
```
