Filen `corpus_tools.py` inneholder verktøy for å utføre kvantive analyser av diktkorpus.

```python
# Importer funksjoner 
from corpus_tools import (get_poems, # Henter dikt fra database
                          concordance, # Finner konkordanser
                          collocation, # Finner kollokasjoner
                          total_word_count, # Frekvensliste for tekstsamling
                          get_dtm,  # Frekvensliste per tekst
                          count_tokens, # Telle flere ord
                          count_token, # Tell ett ord
                          relative_collocation, # Finn relativ konkordans
                          get_reference_corpus # Hent referansekorpus fra dhlab
                          )
```

```python
# Hent dikt fra database
poems = get_poems()
```

```python
# Denne funksjonen teller antall forekomster av utvalgte ord
count_tokens(["fjell", "fjord"], poems)
```

```python
# Totalt antall ord i diktene
tot = total_word_count(poems)
tot
```

```python
# Totalt antall ord
tot.sum()
```

```python
# Konkordanse for ordet "fjell"
concordance("fjell", poems, window=25).head()
```

```python
# Hent dokumentasjonsstreng for funksjonen
??concordance
```

```python
# Alternativ visning av konkordanser
# Style metdoen gir en penere visning av datarammen
concordance("fjord", poems, window=25, join=True).head(10).style
```

```python
# Absolutte kollokasjoner av ordet "fjell"
collocation("fjell", poems, window=25).head(10)
```

```python
# relative kollokasjoner av ordet "fjell"
tot = total_word_count(poems)
```

```python
count_token("1851", poems)
```

```python
# Relativ konkordans for ordet "fjell"
relative_collocation("fjell", poems, tot, window=25).head(30)
```

```python
# Relativ frekvens for ordet "fjell"
(collocation("fjell", poems) / tot).sort_values(by="frequency", ascending=False).head(20)
```

```python
# Hent de 50 000 vanligste ord fra norsk 1800-tall (bøker)
ref_corpus = get_reference_corpus()
```

```python
# Sammenlign kollokasjoner i fjell med referansekorpuset for norsk 1800-tall
relative_collocation("fjell", poems, ref_corpus, window=25).head(30)
```

```python
# Hent frekvensliste per tekst (dokument-term-matrise)
dtm = get_dtm(poems)
dtm.head()
```
