```python
import requests


def get_reference_corpus(period=(1800,1900)):
    url = "https://api.nb.no/dhlab/#/default/get_reference_corpus"
    from_year = period[0]
    to_year = period[1]
    res = requests.get(url, params={'from_year':from_year, 'to_year':to_year})
    #if res.response_code == 200:
    return res
    
```
